```python
#reading csv file with pandas
phone = pd.read_csv("phone_data_raw.csv")
phone
```

```python
#covert the startTime and endTime to actual data 
phone['startTime'] = pd.to_datetime(phone['startTime'], errors='coerce')
phone['endTime'] = pd.to_datetime(phone['endTime'], errors='coerce')
phone
```

```python
# creating a variable date and time from endTime 
phone['date'] = phone['startTime'].dt.date
# phone['time'] = phone['endTime'].dt.time
phone['time'] = phone['startTime'].dt.floor('min').dt.time
phone
```

```python
#creating variable time_spend in minutes on the app 
phone['time_spend'] = (phone['endTime'] - phone['startTime'])/np.timedelta64(1,'m')
phone['time_spend'] = round(phone['time_spend'],1)
phone
```

```python
#create a period variable from time 
phone['period'] = (phone['startTime'].dt.hour % 24 + 4) // 4
phone['period'].replace({1: 'Late-night',
                      2: 'Early Morning',
                      3: 'Morning',
                      4: 'Noon',
                      5: 'Evening',
                      6: 'Night'}, inplace=True)
phone.head(80)
```

```python
#creating the day of the week variable
phone['day_of_week'] = phone['startTime'].dt.day_name()
phone
```

```python
#creating a weekend variable 
def category_to_numeric(x):
    if x == 'Monday':
        return 0 
    if x == 'Tuesday':
        return 0 
    if x == 'Wednesday':
        return 0
    if x == 'Thursday':
        return 0 
    if x == 'Friday':
        return 0 
    if x == 'Saturday':
        return 1 
    if x == 'Sunday':
        return 1 
phone['weekend'] = phone['day_of_week'].apply(category_to_numeric)
phone
```

```python
print(phone['date'].max())
print(phone['date'].min())
```

```python
#selecting variables from the dataframe 
application_data = phone.filter(['UserID','session','application', 'time_spend' , 'time','date','day_of_week','weekend', 'period' ], axis=1)

#filter apps that are not ethica logger
application_data = application_data[application_data['application'] != 'com.ethica.logger']
application_data
```

```python
#filter the activities that lasted more than 0 second
application_data = application_data[application_data['time_spend'] > 0.0]
application_data
```

```python
app_id = application_data
app_id.rename(columns = {'application':'app_id'}, inplace = True)
app_id
```

```python
#extracting the UserID of the table 
users = application_data['UserID'].unique()
users = pd.DataFrame(users)
users = users.rename(columns = {0: 'UserID'})
users
```

```python
#creating a new variable total_sessions, a count of total sessions 
total_sessions = application_data.groupby(['UserID'])['session'].count()
total_sessions = pd.DataFrame(total_sessions)
total_sessions = total_sessions.rename(columns = {'session': 'total_sessions'})
total_sessions
```

```python
#creating a social media frame 
sns_data = pd.merge(users, total_sessions, on = 'UserID', how ='left')
sns_data
```

```python
#counting total sessions time 
time_sessions = round(application_data.groupby(['UserID'])['time_spend'].sum(),1)
time_sessions = pd.DataFrame(time_sessions)
time_sessions = time_sessions.rename(columns = {'time_spend': 'time_sessions'})
time_sessions
```

```python
sns_data = pd.merge(sns_data, time_sessions, on = 'UserID', how ='left')
sns_data
```
